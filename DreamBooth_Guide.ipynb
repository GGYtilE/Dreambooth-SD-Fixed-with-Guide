```python
!python main.py --base configs/stable-diffusion/v1-finetune_unfrozen.yaml -t \
                --actual_resume weights/sd-v1-4-full-ema.ckpt \
                -n "red_backpack" \
                --gpus 0, \
                --data_root train_img \
                --reg_data_root outputs/txt2img-samples/samples \
                --class_word "backpack"
```
